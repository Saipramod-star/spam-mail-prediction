```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

```python
raw_mail_data = pd.read_csv(r"C:\Users\SAI RAM\Desktop\ML things\spamham.csv")
raw_mail_data
```

```python
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)), '')
```

```python
mail_data.shape
```

```python
mail_data.head()
```

```python
mail_data.loc[mail_data['Category'] == "spam",'Category',] = 0
mail_data.loc[mail_data['Category'] == "ham",'Category',] = 1
```

```python
mail_data.head()
```

```python
x = mail_data['Message']
y = mail_data['Category']
```

```python
print(x)
print('........')
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.1,random_state = 3)
```

### Feature Extraction

```python
# converting  the text data to feature words that can be used as input to the SVM model using TfidfVectorizer
# convert the text into lower case letters
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase='True')
x_train_features = feature_extraction.fit_transform(x_train)
x_test_features = feature_extraction.fit_transform(x_test)

#convert Y_train and Y_test values as integers
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

### Training the model

```python
model = LinearSVC()
model.fit(x_train_features,y_train)
```

```python
prediction_on_training_data = model.predict(x_train_features)
accuracy_on_training_data = accuracy_score(y_train, prediction_on_training_data)
```

```python
print("Accuracy", accuracy_on_training_data)
```
